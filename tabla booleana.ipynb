```python
booleanos = [False, True]  # Sea 0 para False, y 1 para True

#tabla de verdad de not
print('x\tnot x')
print('-'*15)
for x in booleanos:
    print(x, not x, sep='\t')

print() 


#Tabla de verdad de or

print('x\ty\tx or y')  #\t es tabulaci√≥n
print('-'*22)  

for x in booleanos:
    for y in booleanos:
        print(x, y, x or y, sep = '\t')

print()        


#tabla de verdad de and
print('x\ty\tx and y')
print('-'*22)

for x in booleanos:
    for y in booleanos:
        print(x, y, x and y, sep='\t')

print()    

   
#tabla de verdad de nand
print('x\t y\tx and y\t x nand y')
print('-'*35)

for x in booleanos:
    for y in booleanos:
        print(x, y, x and y, not(x and y), sep='\t ')

print()  

#tabla de verda de condicional (si x entonces y == not x or y)
print('x\t y\t not x or y')
print('-'*30)

for x in booleanos:
    for y in booleanos:
        print(x, y, not x or y, sep='\t')

print()


#Tabla con 3 valores (x, y, z)
print('x\t y\t z\t x and y and z')
print('-'*45)

for x in booleanos:
    for y in booleanos:
        for z in booleanos:
            print(x, y, z, x and y and z, sep='\t')

#print()
```
